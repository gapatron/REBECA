```python
import os
import shutil
```

```python
REBECA_T2_DIR = "../data/flickr/evaluation/diffusion_priors/models/samples/final/SPU_80_GS_10.0_Realistic_image,_finely_detail_NEG_deformed,_ugly,_wrong_proporti_usrthres_100_v1"
REBECA_T0_GLOBAL_DIR = f"{REBECA_T2_DIR}_global_images"
os.makedirs(REBECA_T0_GLOBAL_DIR)
for user in os.listdir(REBECA_T2_DIR):
    #print(user)
    src = f"{REBECA_T2_DIR}/{user}/images"
    dst = f"{REBECA_T0_GLOBAL_DIR}"
    for root, subdirs, files in os.walk(src):
        #print(root)
        for file in files:
            #print(files)
            #print(file)
            path = f"{root}/{file}"
            local_dst = f"{dst}/{user}_{file}"
            #print(path, "\n", local_dst)
            shutil.copy(path, local_dst)
```

```python
import pandas as pd

test_users = pd.read_csv("../data/flickr/evaluation/ground_truth/usrthrs_100/test_usrthrs_100.csv")
```

```python
test_users["rating"] = test_users["score"]>3
test_users
```

```python
import torch

data = torch.load("../data/flickr/processed/test/test_ie_usrthrs_100.pth")
```

```python
import os
savedir = "../data/flickr/evaluation/ground_truth/usrthrs_100/users/"
for user in test_users["worker_id"].unique():
    test_user_df = test_users[test_users["worker_id"] == user]
    user_embs = data[test_user_df.index]
    ratings = torch.tensor(test_user_df["rating"].values).long()
    embs_savepath = os.path.join(savedir, str(user), "embeddings.pt")
    ratings_savepath = os.path.join(savedir, str(user), "ratings.pt")
    os.makedirs(savedir+"/"+str(user), exist_ok=True)
    torch.save(user_embs, embs_savepath)
    torch.save(ratings, ratings_savepath)

    
```
